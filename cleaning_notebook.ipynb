```python


def clean_tide_max(df):
    df = df.drop(columns= ['Predicted (ft)', 'Preliminary (ft)'], axis=1)
    df['Date'] = pd.to_datetime(df['Date'])
    df = df.set_index('Date')
    #df.index = pd.to_datetime(df.index, format='%Y-%m')
    df = df[df['Verified (ft)'] != '-']
    df['Verified (ft) Max'] = pd.to_numeric(df['Verified (ft)'])
    df = df.resample('D').max()
    return df
```

```python
def clean_unemploy(df, noaa_data):
    df.columns = ['Date', 'unemployment_rate']
    df['Date'] = df['Date'].map(lambda x: x[:7])
    df['Date'] = pd.to_datetime(df['Date'])
    df = df.set_index('Date')
    df['unemployment_rate'] = pd.to_numeric(df['unemployment_rate'])
    list_days = noaa_data.index
    rate_list = [df.loc[str(i)[:7]]['unemployment_rate'].item() for i in list_days]
    unemploy_df = pd.DataFrame(data=rate_list, index=list_days)
    unemploy_df.rename(columns={0: 'rate'}, inplace=True)
    return unemploy_df
```

```python
from statsmodels.tsa.stattools import adfuller

# Code written by Joseph Nelson.

def interpret_dftest(dftest):
    dfoutput = pd.Series(dftest[0:2], index=['Test Statistic','p-value'])
    return dfoutput

interpret_dftest(adfuller(boston_max['financial_activities_avg_monthly_pay']))
```

```python
# Write a for loop to find the lowest difference value d.

# Iterate through all possible differences.
for d in range(1, len(boston_max['financial_activities_avg_monthly_pay'])):
    
    # Print out a counter and the corresponding p-value.
    print(f'Checking difference of {d}.')
    print(f'p-value = {interpret_dftest(adfuller(boston_max["financial_activities_avg_monthly_pay"].diff(d).dropna()))["p-value"]}.')

          
    # If our data, differenced by d time periods, are stationary, print that out!
    if interpret_dftest(adfuller(boston_max['financial_activities_avg_monthly_pay'].diff(d).dropna()))['p-value'] < 0.05:
        print(f'Differencing our time series by d={d} yields a stationary time series!')
        break
    
    print()
```
